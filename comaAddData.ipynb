```python
import pandas as pd
import os

config_files = ['D:\PLFS\StandardPLFSUpdateSheet2017-2018.xlsx','D:\PLFS\StandardPLFSUpdateSheet2018-2019.xlsx','D:\PLFS\StandardPLFSUpdateSheet2019-2020.xlsx'
               ,'D:\PLFS\StandardPLFSUpdateSheet2020-2021.xlsx','D:\PLFS\StandardPLFSUpdateSheet2021-2022.xlsx','D:\PLFS\StandardPLFSUpdateSheet2022-2023.xlsx']

for file in config_files:
    print('Shet :-' + file)
    sheets_dict = pd.read_excel(file, sheet_name=None)
    sheet_names = sheets_dict.keys()
    print(sheet_names)
    sheet_names = [name for name in sheets_dict.keys() if '(' not in name and ')' not in name]

    sheet_names_df = pd.DataFrame(sheet_names, columns=['Sheet Names'])
    # Define the path to the new Excel file
    new_file_path = r'D:\PLFS\SheetName'  # Use raw string to avoid issues with backslashes
    
    # Ensure the directory exists
    os.makedirs(new_file_path, exist_ok=True)
    
    # Construct the output file path
    output_file = os.path.join(new_file_path, os.path.basename(file))
    
    # Write the DataFrame to a new Excel file
    sheet_names_df.to_excel(output_file, index=False)
    print(f'Sheet names saved to: {output_file}')
```

```python
options = ["Female", "Male", "All", "Person"]

# Define the custom order
custom_order = ["Male", "Female", "Person", "All"]

# Create a dictionary to map each option to its index in the custom order
order_dict = {value: index for index, value in enumerate(custom_order)}

# Sort the options list based on the custom order
sorted_options = sorted(options, key=lambda x: order_dict.get(x, float('inf')))

print(sorted_options)
```
