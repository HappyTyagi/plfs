```python
for key in df2:
    df3 =  df2[key]
    # df3 = df3.str.capitalize()
    df3 = df3.replace("-" , np.nan)
    # df3 = sperator(df3)
    df3 = df3.drop_duplicates()
    df3 = df3.dropna()

    df3 = df3.reset_index(drop = True)
    
    name = ""
    for character in key:
        if character not in ["(" , ")" , "/" , "#"]:
            name+= character
    print(df3)         
    df3.to_excel(f"PLFS_MASTER_EXCEL_SHEETS\{name.split(' ')[0]}.xlsx" , index = False)
    # print(name.split(" ")[0])
    print(f"{key} done")
```
